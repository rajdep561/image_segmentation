```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Paths to your dataset where images are organized in class subdirectories
train_data_dir = '/home/rajdeep/projects/image_segmentation/data/archive/img/'
output_dir = '/home/rajdeep/projects/image_segmentation/data/archive/augmented'
class_names = np.unique(df.loc[:,['label']])#os.listdir(train_data_dir)

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Define the target number of images per class
target_num_images = 1000

# Initialize the ImageDataGenerator with your desired augmentations
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

for class_name in class_names:
    class_dir = os.path.join(train_data_dir, class_name)
    output_class_dir = os.path.join(output_dir, class_name)
    os.makedirs(output_class_dir, exist_ok=True)
    
    # List all images in the class directory
    image_files = [f for f in os.listdir(class_dir) if f.endswith('.jpg')]

    # Calculate how many additional images are needed
    num_existing_images = len(image_files)
    num_images_needed = target_num_images - num_existing_images
    
    if num_images_needed > 0:
        print(f"Augmenting class '{class_name}' with {num_images_needed} images.")
        
        # Load and augment existing images
        for img_file in image_files:
            img_path = os.path.join(class_dir, img_file)
            img = tf.keras.preprocessing.image.load_img(img_path)
            img = tf.keras.preprocessing.image.img_to_array(img)
            img = np.expand_dims(img, axis=0)

            # Generate augmented images and save them
            i = 0
            for batch in datagen.flow(img, batch_size=1, save_to_dir=output_class_dir, save_prefix=class_name, save_format='jpeg'):
                i += 1
                if i >= num_images_needed:
                    break
    else:
        print(f"Class '{class_name}' already has enough images.")
```
