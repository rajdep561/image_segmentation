```python
import os
import pandas as pd
import numpy as np
import os
import seaborn as sns
from PIL import Image
import matplotlib.pyplot as plt
def read_images(folder_path,kv=False):
    
    if kv==False:
        image_list = []
        raw_image = []
        for filename in os.listdir(folder_path):
            if filename.endswith('.jpg') or filename.endswith('.png'):  # Adjust file extensions as needed
                image_path = os.path.join(folder_path, filename)

                image_list.append(image_path)
    if kv==True:
        image_list={}
        for filename in os.listdir(folder_path):
            if filename.endswith('.jpg') or filename.endswith('.png'):  # Adjust file extensions as needed
                image_path = os.path.join(folder_path, filename)
                image_list[filename.replace('_segmentation','').split('.')[0]] = image_path
                
    return image_list
   
        

# Example usage
directory_path = '/home/rajdeep/projects/image_segmentation/data/segmented_untransformed/'  # Replace with the path to your directory
image_list = read_images(directory_path)
# os.mkdir('./data/preprocessed_masks')
# os.mkdir('./data/preprocessed_masks/transformed')
# os.mkdir('./data/preprocessed_masks/untransformed')
```

```python
def uniformity(image_path):
    img_name = image_path.split('/')[-1]
    image = Image.open(image_path)
    image_array = np.array(image)
    height, width = image_array.shape
    central_percentage = 0.35
    central_width = int(width * central_percentage)
    central_height = int(height * central_percentage)

    start_x = (width - central_width) // 2
    start_y = (height - central_height) // 2

    end_x = start_x + central_width
    end_y = start_y + central_height
    
    
    central_region = image_array[start_y:end_y, start_x:end_x]
    central_pixel_value = np.mean(central_region)
#     print(central_pixel_value)
    # Access the center pixel value
#     center_pixel_value = image_array[center_row, center_col]
    if central_pixel_value < 10:
        max_pixel_value = 255
        complement_image = max_pixel_value - image_array

        im = Image.fromarray(complement_image)
        im.save('./data/preprocessed_masks/untransformed/'+img_name)
    else:
        im = Image.fromarray(image_array)
        im.save('./data/preprocessed_masks/untransformed/'+img_name)
    
```
